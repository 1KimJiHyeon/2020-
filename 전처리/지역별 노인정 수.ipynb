```
location = df['관할행정동'].str.lstrip()
```

```
location
```

```
location_list = list(location)
```

```
location_dict = {}
prev_location=""
count = 1

for i in range(len(location_list)):
    if i==0:
        prev_location="의창구 동읍 "
    else:
        if prev_location == location_list[i]:
            count = count + 1
        else:
            location_dict[prev_location] = count
            count=1
        prev_location = location_list[i]
        
location_dict[prev_location] = count
```

```
location_dict
```

```
#dictionary to dataFrame
total_df = pd.DataFrame.from_dict(location_dict,orient='index')
total_df.columns = ['노인정 수']
total_df
```

```
total_df['구/동읍면'] = total_df.index
total_df = total_df[['구/동읍면','노인정 수']]
total_df=total_df.reset_index(drop=True)
total_df
```

```
address = pd.read_excel('new_adress.xlsx')
address
location_count = total_df['구/동읍면'].str.split(' ').str[1]
location_count= list(location_count)
address = list(address['읍면동명'])
address
```

```
isExist= False
additional_list = []
for i in location_count:
  for j in address:
    if i == j:
      isExist = True
  
  if isExist == False:
    additional_list.append(i)
  isExist =  False
```

```
additional_list
```

```
for i in additional_list:
  if "중앙동" in i:
    additional_list.remove(i)
additional_list
```

```
#마산합포구 노산동  -> 마산합포구 교방동 +8 12
#진해구 태평동 - > 진해구 충무동 +9 18
delete_nm1 = total_df[total_df['구/동읍면'] == "마산합포구 노산동"].index
total_df = total_df.drop(delete_nm1)
delete_nm3 = total_df[total_df['구/동읍면'] == "마산합포구 교방동"].index
total_df = total_df.drop(delete_nm3)
print(delete_nm1)


delete_nm2 = total_df[total_df['구/동읍면'] == "진해구 태평동"].index
total_df = total_df.drop(delete_nm2)
delete_nm4 = total_df[total_df['구/동읍면'] == "진해구 충무동"].index
total_df = total_df.drop(delete_nm4)
delete_nm5= total_df[total_df['구/동읍면'] == "진해구 중앙동"].index
total_df = total_df.drop(delete_nm5)

additional_df = pd.DataFrame([['마산합포구 교방동',12],['진해구 충무동',21]],columns=['구/동읍면','노인정 수'])
total_df=total_df.append(additional_df, ignore_index=True)
```
