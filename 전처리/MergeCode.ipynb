```
merged = pd.read_csv('merged.csv')
merged
```

```
# 읍면동명 참고 version
code_list = []

for i in range(merged.shape[0]):
  merged_addr = merged.iloc[i]['동']
  for j in range(Changwon_code.shape[0]):
    code_addr = Changwon_code.iloc[j]['읍면동명']
    if merged_addr == code_addr:
      code_list.append(Changwon_code.iloc[j]['행정동코드'])
      break

print(code_list)
# len(code_list)
```

```
merged['Code'] = code_list
merged
```
