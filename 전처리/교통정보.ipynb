```
# 구, 동/읍/면/리 column추가
df['구'] = df.주소.str.split(' ').str[0]
df['동'] = df.주소.str.split(' ').str[1]
df
```

```
# 주소 변경
df.replace(['장천동', '성내동', '마천동', '용원동', '부림동', '평화동', '대내동', '회원동', '석전동', '소답동', '사림동', '신월동', '반지동', '대방동']
            , ['풍호동', '웅천동', '웅동1동', '웅동2동', '오동동', '문화동', '월영동', '회원1동', '석전1동', '의창동', '봉림동', '용지동', '봉림동', '사파동'])

df
```

```
# 구, 동 로 그룹
grouped = df.groupby(['구', '동'])

# result = 최종 결과값
result = grouped.count().reset_index()
result.columns = ['구', '동', '버스정보안내기 수']
result
```
