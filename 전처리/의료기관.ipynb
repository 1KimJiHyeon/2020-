```
care_changwon_df = pd.read_csv('care_changwon.csv',encoding="euc-kr")
```

```
care_masan_df = pd.read_csv('care_masan.csv',encoding="euc-kr")
```

```
care_jinhae_df = pd.read_csv('care_jinhae.csv',encoding="euc-kr")
```

```
address_df = pd.read_excel('address.xlsx',sheet_name='Sheet1')
```

```
care_changwon_df.drop(['순번','의료기관종별','의료기관주소(도로명)'], axis=1, inplace=True)
```

```
care_masan_df.drop(['의료기관전화번호','의료기관종별','의료기관주소(도로명)'], axis=1, inplace=True)
```

```
care_jinhae_df.drop(['종별','소재지','전화번호'], axis=1, inplace=True)
```

```
care_df = pd.concat([care_changwon_df,care_masan_df,care_jinhae_df], ignore_index=True)
care_df
```

```
care_address = list(care_df['읍면동'])
care_address
```

```
address_list=list(address_df['동읍면'])
address_list
```

```
old_dong_list=[]

for i in care_address:
    check=False
    for k in address_list:
        if i==k:
            check=True
            break
    if check == False:
        old_dong_list.append(i)

old_dong_set = set(old_dong_list)
old_dong_set
```

```
gu_list=[]
dong_list=[]
gu_count_list=[]
dong_count_list=[]

for i in (list(set(address_df['구']))):
    care_gu_count=0
    for j in (list(address_df[address_df['구']==i]['동읍면'])):
        care_dong_count=0
        for k in care_address:
            if j==k:
                care_gu_count+=1
                care_dong_count+=1
                        
        gu_list.append(i)
        dong_list.append(j)
        dong_count_list.append(care_dong_count)

    for k in range (len(list(address_df[address_df['구']==i]['동읍면']))):
        gu_count_list.append(care_gu_count)
```

```
data = {"구":gu_list,"구별 의료기관 수":gu_count_list, "동":dong_list,"동별 의료기관 수":dong_count_list}
care_DataFrame = pd.DataFrame(data)
care_DataFrame
```

```
care_DataFrame.to_csv('care_DataFrame.csv',encoding = 'UTF-8')
```
